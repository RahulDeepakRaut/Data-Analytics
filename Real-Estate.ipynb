# 📊 Property & Pricing Analysis

```python
# 1)What is the average listing price by city?

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql369",
    database="real_estate"
    )

print(con)

cursor = con.cursor(buffered=True)

query = """
SELECT 
City,
ROUND(AVG(Price), 2) AS Avg_Listing_Price
FROM listings
GROUP BY City
ORDER BY Avg_Listing_Price DESC
"""

cursor.execute(query)
results = cursor.fetchall()

print("\nAverage Listing Price by City:")
print("-" * 40)
print(f"{'City':<20} {'Avg Listing Price':>20}")
print("-" * 40)
for row in results:
    print(f"{row[0]:<20} {row[1]:>19,.2f}")
    print("-" * 40)

cursor.close()
con.close()
```

```python
# 2) What is the average price per square foot by property type?


con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql369",
    database="real_estate"
    )

print(con)

cursor = con.cursor(buffered=True)

query = """
        SELECT 
            Property_Type,
            ROUND(AVG(Price / Sqft), 2) AS Avg_Price_Per_Sqft
        FROM listings
        WHERE Sqft > 0
        GROUP BY Property_Type
        ORDER BY Avg_Price_Per_Sqft DESC
    """

cursor.execute(query)
results = cursor.fetchall()

print("\nAverage Price per Square Foot by Property Type:")
print("-" * 50)
print(f"{'Property Type':<20} {'Avg Price Per Sqft':>25}")
print("-" * 50)
for row in results:
    print(f"{row[0]:<20} ${row[1]:>24,.2f}")
    print("-" * 50)

cursor.close()
con.close()
```

```python
# 3) How does furnishing status impact property prices?

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql369",
    database="real_estate"
    )

print(con)

cursor = con.cursor(buffered=True)

query = """
        SELECT 
            pa.furnishing_status,
            COUNT(l.Listing_ID)             AS Total_Listings,
            ROUND(AVG(l.Price), 2)          AS Avg_Price,
            ROUND(MIN(l.Price), 2)          AS Min_Price,
            ROUND(MAX(l.Price), 2)          AS Max_Price,
            ROUND(AVG(l.Price / l.Sqft), 2) AS Avg_Price_Per_Sqft
        FROM listings l
        JOIN property_attributes pa
        ON l.Listing_ID = pa.listing_id
        WHERE l.Sqft > 0
        GROUP BY pa.furnishing_status
        ORDER BY Avg_Price DESC
    """

cursor.execute(query)

results = cursor.fetchall()

print("\nImpact of Furnishing Status on Property Prices:")
print("-" * 90)
print(f"{'Furnishing Status':<20} {'Total Listings':>15} {'Avg Price':>18} {'Min Price':>18} {'Max Price':>18}")
print("-" * 90)
for row in results:
    print(f"{row[0]:<20} {row[1]:>15,} ${row[2]:>17,.2f} ${row[3]:>17,.2f} ${row[4]:>17,.2f}")
    print("-" * 90)

cursor.close()
con.close()
```

```python
# 4) Do properties closer to metro stations command higher prices?

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql369",
    database="real_estate"
    )

print(con)

cursor = con.cursor(buffered=True)

query = """
        SELECT 
            CASE 
                WHEN pa.metro_distance_km <= 2  THEN '0-2 km (Very Close)'
                WHEN pa.metro_distance_km <= 5  THEN '2-5 km (Close)'
                WHEN pa.metro_distance_km <= 8  THEN '5-8 km (Moderate)'
                WHEN pa.metro_distance_km <= 10 THEN '8-10 km (Far)'
                ELSE 'Above 10 km (Very Far)'
            END                                   AS Metro_Distance_Range,
            COUNT(l.Listing_ID)                   AS Total_Listings,
            ROUND(AVG(l.Price), 2)                AS Avg_Price,
            ROUND(MIN(l.Price), 2)                AS Min_Price,
            ROUND(MAX(l.Price), 2)                AS Max_Price,
            ROUND(AVG(l.Price / l.Sqft), 2)       AS Avg_Price_Per_Sqft,
            ROUND(AVG(pa.metro_distance_km), 2)   AS Avg_Metro_Distance_km
        FROM listings l
        JOIN property_attributes pa
        ON l.Listing_ID = pa.listing_id
        WHERE l.Sqft > 0
        GROUP BY Metro_Distance_Range
        ORDER BY Avg_Metro_Distance_km ASC
    """

cursor.execute(query)
results = cursor.fetchall()

print("\nImpact of Metro Distance on Property Prices:")
print("-" * 110)
print(f"{'Distance Range':<25} {'Listings':>10} {'Avg Price':>18} {'Min Price':>18} {'Max Price':>18} {'Avg ₹/Sqft':>15}")
print("-" * 110)
for row in results:
    print(f"{row[0]:<25} {row[1]:>10,} ${row[2]:>17,.2f} ${row[3]:>17,.2f} ${row[4]:>17,.2f} ${row[5]:>14,.2f}")
    print("-" * 110)

cursor.close()
con.close()
```

```python
# 5) Are rented properties priced differently from non-rented ones?

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql369",
    database="real_estate"
    )

print(con)

cursor = con.cursor(buffered=True)

query = """
        SELECT 
            CASE 
                WHEN pa.is_rented = 1 THEN 'Rented'
                ELSE 'Not Rented'
            END                                   AS Rental_Status,
            COUNT(l.Listing_ID)                   AS Total_Listings,
            ROUND(AVG(l.Price), 2)                AS Avg_Price,
            ROUND(MIN(l.Price), 2)                AS Min_Price,
            ROUND(MAX(l.Price), 2)                AS Max_Price,
            ROUND(AVG(l.Price / l.Sqft), 2)       AS Avg_Price_Per_Sqft,
            ROUND(AVG(l.Sqft), 2)                 AS Avg_Sqft,
            ROUND(AVG(pa.tenant_count), 2)        AS Avg_Tenant_Count,
            ROUND(AVG(pa.metro_distance_km), 2)   AS Avg_Metro_Distance_km
        FROM listings l
        JOIN property_attributes pa ON l.Listing_ID = pa.listing_id
        WHERE l.Sqft > 0
        GROUP BY pa.is_rented
        ORDER BY Avg_Price DESC
    """

cursor.execute(query)
results = cursor.fetchall()

print("\nRented vs Non-Rented Property Price Comparison:")
print("-" * 110)
print(f"{'Rental Status':<15} {'Listings':>10} {'Avg Price':>18} {'Min Price':>18} {'Max Price':>18} {'Avg/Sqft':>12} {'Avg Sqft':>10}")
print("-" * 110)
for row in results:
    print(f"{row[0]:<15} {row[1]:>10,} ${row[2]:>17,.2f} ${row[3]:>17,.2f} ${row[4]:>17,.2f} ${row[5]:>11,.2f} {row[6]:>10,.2f}")
    print("-" * 110)

cursor.close()
con.close()
```

```python
# 22) Does experience correlate with deals closed?

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql369",
    database="real_estate"
    )

print(con)

cursor = con.cursor(buffered=True)

query1 = """
    SELECT 
        CASE 
            WHEN a.experience_years <= 5 THEN '0-5 Yrs (Junior)'
            WHEN a.experience_years BETWEEN 6 AND 10 THEN '6-10 Yrs (Mid-Level)'
            WHEN a.experience_years BETWEEN 11 AND 15 THEN '11-15 Yrs (Senior)'
            ELSE '16+ Yrs (Veteran)'
        END AS Experience_Tier,
        COUNT(DISTINCT a.Agent_ID)             AS Total_Agents,
        ROUND(AVG(a.deals_closed), 1)          AS Avg_Lifetime_Deals,
        ROUND(AVG(a.rating), 2)                AS Avg_Rating,
        COUNT(s.Listing_ID)                    AS Total_Current_Sales,
        ROUND(COUNT(s.Listing_ID) / COUNT(DISTINCT a.Agent_ID), 1) AS Avg_Sales_Per_Agent
    FROM agents a
    LEFT JOIN listings l ON a.Agent_ID = l.Agent_ID
    LEFT JOIN sales s ON l.Listing_ID = s.Listing_ID
    GROUP BY Experience_Tier
    ORDER BY MIN(a.experience_years) ASC
"""

cursor.execute(query1)
results1 = cursor.fetchall()

print(f"\nPerformance Metrics by Experience Cohorts: {len(results1)}")
print("-" * 110)
print(f"{'Experience Tier':<22} {'Agent Count':>12} {'Avg Lifetime Deals':>20} {'Avg Rating':>12} {'Total Dataset Sales':>21} {'Avg Sales/Agent':>16}")
print("-" * 110)
for row in results1:
    print(f"{row[0]:<22} {row[1]:>12,} {row[2]:>20,} {row[3]:>12,} {row[4]:>21,} {row[5]:>16,}")
print("-" * 110)

cursor.close()
con.close()
```
